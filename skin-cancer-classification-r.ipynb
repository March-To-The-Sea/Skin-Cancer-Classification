```r
non_fraud <- length(which(CSV_READ$Class == 0))
fraud <- length(which(CSV_READ$Class == 1))
fraud_percent <- (fraud / (fraud + non_fraud)) * 100

cat("\n Number of Genuine transactions: ", non_fraud)
cat("\n Number of Fraud transactions: ", fraud)
cat("\n Percentage of Fraud transactions: ", format(round(fraud_percent, 4), n_small = 3))
```
